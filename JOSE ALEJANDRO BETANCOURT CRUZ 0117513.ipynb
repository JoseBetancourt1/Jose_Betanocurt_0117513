```python

"""
EJERCICIO 6

Leer un número entero y determinar si la suma de sus dígitos es un número 
de Fibonacci.
"""
def fib(num):
    pre=0
    pos=1
    bandera=False
    
    while(pre<=num):
        
        if (pre == num):
            bandera=True
        
        suma= pre+pos
        pre= pos
        pos=suma
    
    return bandera
num= int(input('Ingrese numero: '))

if (fib(num)):
    print('El numero pertenece a la serie Fibonacci')
else:
    print('El numero NO pertenece a la serie Fibonacci')
    
```

```python
"""
EJERCICIO 9

Haga un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) 
que corresponden al centro y al radio de dos círculos. Lea un punto de 
coordenadas (a,b). Determina si (a,b) está contenido: 
    a) dentro del círculo 1; 
    b) dentro del círculo 2; 
    c) dentro de ambos círculos; 
    d) fuera de ambos círculos. 
Recordemos que el círculo es el lugar geométrico de los puntos del plano cuya 
distancia a otro punto fijo, llamado centro, es menor o igual que una cantidad 
constante, llamada radio.
"""

def punto_pertenece_a_dos_cir(x1,y1,r1,x2,y2,r2,a,b):
    #formula de la circunferencia (a-x)**2+(b-y)**2=r**2
    #centro en(x,y)
    per_cir=(a-x1)**2+(b-y1)**2 ; per_cir_2=(a-x2)**2+(b-y2)**2
    rad1=r1**2                      ; rad2=r2**2
    
    return per_cir,rad1,rad2,per_cir_2
while True:
    try:
        X1=float(input("Ingrese la coordenada en X del primer circulo : "))
        Y1=float(input("Ingrese la coordenada en Y del primer circulo : "))
        R1=float(input("Ingrese el radio de la primera circunferencia : "))
        X2=float(input("Ingrese la coordenada en X del segundo circulo : "))
        Y2=float(input("Ingrese la coordenada en Y del segundo circulo : "))
        R2=float(input("Ingrese el radio de la circunferencia : "))
        A=float(input("Ingrese la coordenada en X a calcular : "))
        B=float(input("Ingrese la coordenada en Y a calcular : "))
        break
    except ValueError:
        print("El caracter ingresado es incorrecto. Inténtelo de nuevo.")

per_cir,rad1,rad2,per_cir_2=punto_pertenece_a_dos_cir(X1,Y1,R1,X2,Y2,R2,A,B)
if per_cir<=rad1 and per_cir<=rad2:
    print("El punto pertenece a las dos circulos.")
elif per_cir<=rad1:
    print("El punto pertenece al circulo no.1")
elif per_cir_2<=rad2:
    print("El punto pertenece al circulo no.2")
else:
    print("El punto no pertenece a ningun circulo.")
```

```python
"""
EJERCICIO 11

Escribir un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.
Utilizar para ello los métodos de cadenas especializados como .isupper(), 
.isdigit(), etc. y la palabra reservada in. 
"""
from keyword import kwlist
cont_voc_may= 0
cont_let_ti = 0
cont_dig = 0
cont_esp = 0
cont_pal_res = 0

cade = input("ingrese una cadena de texto: ")
for i in cade:
    if i in "AEIOUÁÉÍÓÚ":
        cont_voc_may += 1
    if i in "aeiouáéíóú":
        cont_let_ti += 1
    elif i.isdigit():
        cont_dig += 1
    elif i == " ":
        cont_esp += 1
        
for i in tuple(cade.split(" ")):
    if i in kwlist:
        cont_pal_res += 1
print("\nEn la cadena de texto hay: \n "
      "{0} vocal{1} en mayusculas. \n"
      .format(cont_voc_may, "" if cont_voc_may == 1 else "es"),
      "{0} letra{1} con tilde. \n"
      .format(cont_let_ti, "" if cont_let_ti == 1 else "s"),
      "{0} digito{1}. \n"
      .format(cont_dig, "" if cont_dig == 1 else "s"),
       "{0} espacio{1}. \n"
      .format(cont_esp, "" if cont_esp == 1 else "s"),
       "{0} palabra{1} restringida{1}. \n"
      .format(cont_pal_res, "" if cont_pal_res == 1 else "s"))


```

```python
"""
EJERCICIO 13

Escribir un programa que lea una cadena de texto y la imprima como un triángulo
"""
Cd= input('Ingrese una frase: ')
Cd_2=Cd.replace(' ','')
print('\n')

for i in range(1,len(Cd_2)):
    print(Cd_2[(i-1)**2:i**2].center(92))
```
