```python
"""
EJERCICIO 15
Leer una lista de números enteros y enumerar cuantos elementos se repiten 
exactamente dos veces. Utilizar para tal
fin el método count() de listas.
"""


#%%
def Leer_lista():
    while True   :
            try      :
                Lista = eval(input('Ingrese una lista: '))
                if type(Lista) is list:
                    break
                else :
                    print('debe ser una lista [,]')
            except  :
                print('Intentalo de nuevo, debe ser una lista')
    return Lista  
 
#%%
#comprueba que cada elemento sea un entero
def Entero(Lista):
    for i in Lista:
        if type(i) is int:
           C=True
        else:
            C=False
    return C
#%%
def Digitos_numero(Numero):
    
    '''
    '''
    
    Digitos=[]
    while True:
        Dig=Numero%10
        Digitos.append(Dig)
        Numero//=10
        if Numero==0: break
    return Digitos
#%%
        
### evaluar precencia decimal ###            
while True:
    Lista_1=Leer_lista()
    if Entero(Lista_1)== True:
        break
    else:
        print('Los elemtos deben ser numeros enteros,intentelo de nuevo.')

C=0
Numeros_repetidos=[]
for i in Lista_1:
    if i not in Numeros_repetidos:
        Rep=Lista_1.count(i)
        if Rep==2:
            C+=1
            Numeros_repetidos.append(i)
            
if len(Numeros_repetidos)!=0:        
    print(f'Hay {C} digitos que se repiten  en la lista')
else:
    print('No hay digitos repetidos en la lista')
```
